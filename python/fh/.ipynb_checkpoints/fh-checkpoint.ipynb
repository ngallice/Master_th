$\rho=f(h)g(\tau)$ 

$f(h)=h$

$\tau_m \frac{dh}{dt}=-h+I$

$g(\tau)=1-exp(-\gamma\tau)$


```python
t_window=50
dt=0.1

t_final=200
N=int(t_window/dt)
N_sim=int(200/dt)

```

```python
I_initial=1
I_final=4

def I(t):
    if t<0.25*t_final:
        return I_initial
    else:
        return I_final
```

```python
tau_m=20
gamma=0.01
```

```python
def f(h):
    if h<0:
        return 0
    else:
        return h
    
def g(r):
    return (1-np.exp(-gamma*r))
```

```python

q_old=np.zeros((N+1,1))
q_old[0]=1

q=np.zeros((N+1,int(N_sim/10)))

h=I_initial*np.ones((N_sim+1,1))
A=np.zeros((N_sim,1))

q_new=np.zeros((N+1,1))

i_store=0

for it in range(N_sim):
    
    t=it*dt
    
    h[it+1]=(1-dt/tau_m)*h[it]+dt*I(t)/tau_m
    
   
   
    A[it]=q_new[0]#/dt
    q_new = np.zeros((N + 1, 1))
   
    for ir in range(N):
        
        rho =  f(h[it])*g(ir*dt)
       

        q_new[ir + 1] = q_old[ir] * (1 - dt * rho)
        q_new[0] += dt * rho * q_old[ir]

    
    rho = f(h[it])*g(N*dt)
    q_new[N] += q_old[N] * (1 - dt * rho)
    q_new[0] += dt * rho * q_old[N]
    
    #q_new/=(np.sum(q_new*dt))


    q_old = q_new
    
   

    
    
    if it%10==0:
        q[:,i_store]=q_new.flatten()
        i_store+=1
        
```

Calculate L of t

```python
h=I_initial*np.ones((N_sim+1,1))
rho = np.zeros((N+1, N_sim+1))



for it in range(N_sim):
    
    t=it*dt
    
    h[it+1]=(1-dt/tau_m)*h[it]+dt*I(t)/tau_m
    
   
   
    for ir in range(N+1):
        
        rho[ir,it] =  f(h[it])*g(ir*dt)
       

L = np.zeros((N + 1, N + 1,N_sim+1))
for it in range(N_sim+1):
    for i in range(N - 1):
        L[0, i + 1,it] = rho[i,it] * dt
        L[i + 1, i + 1,it] = -1 - rho[i,it] * dt
        L[i + 2, i + 1,it] = 1

    L[0, 0,it] = -1
    L[1, 0,it] = 1
    L[0, N,it] = 1
    L[N, N,it] = -1
```

```python
eigv, vec = np.linalg.eig(L)
eigv = eigv / dt
```

```python
e0=[]
e1=[]
e2=[]
et=[]



for it in range(N_sim):

    t = it * dt

    # Phi


    #drawnow(makeFig)

    if it%50==0:
        eigv, vec = np.linalg.eig(L[:, :, it])
        eigv = eigv / dt

        real_eigv = eigv.real
        x = np.sort(real_eigv)

        eig_0 = np.where(eigv.real == x[-1])
        v_0 = vec[:, eig_0[0][0]].real
        e0.append(x[-1])

        if v_0[0] < 0:
            v_0 = -1 * v_0

        norm = np.sum(v_0) * dt
        v_0 = v_0 / norm

        eig_1 = np.where(eigv.real == x[-2])
        v_1 = vec[:, eig_1[0][0]].real
        e1.append(x[-2])

        if np.max(v_1) != np.max(abs(v_1)):
            v_1 = -1 * v_1

        eig_2 = np.where(eigv.real == x[-4])
        v_2 = vec[:, eig_2[0][0]].real
        e2.append(x[-4])

        if np.max(v_2) != np.max(abs(v_2)):
            v_2 = -1 * v_2


        et.append(t)



```
