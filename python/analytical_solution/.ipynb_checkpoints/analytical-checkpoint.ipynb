```python
t_window=50
dt=0.1
N=int(t_window/dt)
N_sim=int(200/dt)

v=2.
a=0.01

q_old=np.zeros((N+1,1))
q_old[0]=1

q=np.zeros((N+1,int(N_sim/10)))

A=np.zeros((N_sim,1))

q_new=np.zeros((N+1,1))

i_store=0

for it in range(N_sim):
    
    t=it*dt
   
    A[it]=q_new[0]#/dt
    q_new = np.zeros((N + 1, 1))
   

    for ir in range(N):
        

        rho =  v*(1-np.exp(-a*ir*dt))


        q_new[ir + 1] = q_old[ir] * (1 - dt * rho)
        q_new[0] += dt * rho * q_old[ir]

    
    rho = v*(1-np.exp(-a*N*dt))
    q_new[N] += q_old[N] * (1 - dt * rho)
    q_new[0] += dt * rho * q_old[N]
    
    q_new/=(np.sum(q_new*dt))


    q_old = q_new

    
    
    if it%10==0:
        q[:,i_store]=q_new.flatten()
        i_store+=1
        
```

```python
time=-1
t_q=[j*dt for j in range(N+1)]
plt.figure(figsize=(8,6))
plt.plot(t_q, q[:,time],label='q')
plt.xlabel('r [ms]')
plt.ylabel('q')
plt.legend()
plt.xlim(0,t_window)
```

```python
t_A=[(j)*dt for j in range(N_sim)]
plt.figure(figsize=(8,6))
plt.plot(t_A,A,label='A')
plt.xlabel('t [ms]')
plt.ylabel('A')
plt.legend()
```
