## Poisson Process with absolute refractory period

```python
nu=1
d=2

def RHO(tau,nu,d):
    if tau< d:
        return 0 
    else:
        return nu

def S(tau,nu,d):
    if tau <d:
        return 1
    else:
        return np.exp(-nu*(tau-d))
    
def P(tau,nu,d):
    if tau <d:
        return 0
    else:
        return nu*np.exp(-nu*(tau-d))
    
    
```

```python
def Cnm(n,m,nu,d):
    ln=eigenvalue(n,nu,d)
    lm=eigenvalue(m,nu,d)
    
    if n==m:
        return ln*d*(1+0.5*d*(ln+nu))/(nu*(1+(ln+nu)*d)**2)
    else:
        
        return (ln*(nu+lm))/((ln-lm)*nu*(ln+nu)*(1+d*(lm+nu)))
    

def eigenvalue(n,nu,d):
    return (1/d)*sc.special.lambertw(d*nu*np.exp(nu*d), k=n)-nu
```

# Non Constante Input   a_1(t)=U+iV

```python
#REFRACTORINESS
d=2

#TIME PARAMETER
dt=0.001

tau_max=50
N=int(tau_max/dt)

t_max=300
N_sim=int(t_max/dt)
tsim=[i*dt for i in range(N_sim)]


# H PARMETER
h=np.zeros((N_sim+1,1))
h_0=1
h[0]=h_0
tau_m=0.9
omega=0.1
epsilon=2


#NU PARAMETER
beta=0.5
nu_max=2
nu_0=nu_max/2


#MU 
mu_0=0.5
mu_1=6
def mu(t):
    if t<100:
        return  mu_0
    else:
        return  mu_1
    
```

```python
def dnudt(d,i):
    nu=nu_max/(1+np.exp(-beta*(h[i]-h_0)))
    h_prime=1/tau_m*(-h[i]+mu(i*dt)+epsilon*np.cos(omega*i*dt))
    return beta*nu*h_prime/(1+np.exp(beta*(h[i]-h_0)))
```

### RUNGE KUTA

```python
A=np.zeros((N_sim,1))

nu_vec=np.zeros((N_sim+1,1))

U=np.zeros((N_sim+1,1))
U[0]=1

V=np.zeros((N_sim+1,1))
V[0]=0


for i in range(N_sim):
    t=i*dt
    
    h[i+1]=h[i]+dt/tau_m*(-h[i]+mu(i*dt)+epsilon*np.cos(omega*i*dt))
    nu=nu_max/(1+np.exp(-beta*(h[i]-h_0)))
    
    nu_vec[i]=nu
    
    l1=eigenvalue(1,nu,d)
    phi1_0= (nu+l1)/(1+d*(nu+l1))

    
    A_inf=nu/(d*nu+1)
    
    
    f=(l1+(Cnm(1,1,nu,d)+Cnm(1,-1,nu,d))*dnudt(d,i))
    g=(l1+(Cnm(1,1,nu,d)-Cnm(1,-1,nu,d))*dnudt(d,i))
    
    
    c10=Cnm(1,0,nu,d)*dnudt(d,i)
    
    K1=(f.real*U[i]-g.imag*V[i]+c10.real)
    L1=(g.real*V[i]+f.imag*U[i]+c10.imag)
    
    K2=(f.real*(U[i]+0.5*dt*K1)-g.imag*(V[i]+0.5*dt*L1)+c10.real)
    L2=(g.real*(V[i]+0.5*dt*L1)+f.imag*(U[i]+0.5*dt*K1)+c10.imag)
    
    K3=(f.real*(U[i]+0.5*dt*K2)-g.imag*(V[i]+0.5*dt*L2)+c10.real)
    L3=(g.real*(V[i]+0.5*dt*L2)+f.imag*(U[i]+0.5*dt*K2)+c10.imag)
    
    K4=(f.real*(U[i]+dt*K3)-g.imag*(V[i]+dt*L3)+c10.real)
    L4=(g.real*(V[i]+dt*L3)+f.imag*(U[i]+dt*K3)+c10.imag)
    
    
    U[i+1]=U[i]+dt/6*(K1+2*K2+2*K3+K4)
    V[i+1]=V[i]+dt/6*(L1+2*L2+2*L3+L4)
         
    
    A[i]=A_inf+2*(U[i]*phi1_0.real-V[i]*phi1_0.imag)
```

```python
def phih(h):
    return nu_max/(1+np.exp(-beta*(h-h_0)))

def phih_prime(h,hprime):
    return beta*phih(h)*hprime/(1+np.exp(beta*(h-h_0)))
```

```python
#REFRACTORINESS
d=2

#TIME PARAMETER
dt=0.1

tau_max=50
N=int(tau_max/dt)

t_max=300
N_sim=int(t_max/dt)
tsim=[i*dt for i in range(N_sim)]


# H PARMETER
h=np.zeros((N_sim+1,1))
h_0=1
h[0]=h_0
tau_m=0.9
J=10


#NU PARAMETER
beta=0.5
nu_max=1
nu_0=nu_max/2


#MU 
mu_0=-0.5
mu_1=-2
def mu(t):
    if t<100:
        return  mu_0
    else:
        return  mu_1
    

```

```python
A=np.zeros((N_sim,1))

nu_vec=np.zeros((N_sim+1,1))

U=np.zeros((N_sim+1,1))
U[0]=1

V=np.zeros((N_sim+1,1))
V[0]=0

h=np.zeros((N_sim+1,1))
h[0]=h_0


for i in range(N_sim):
    t=i*dt
    
    nu=phih(h[i])
    
    l1=eigenvalue(1,nu,d)
    phi1_0= (nu+l1)/(1+d*(nu+l1))

    A_inf=nu/(d*nu+1)
    
    hprime=-h[i]/tau_m + J/tau_m*(A_inf +2*(U[i]*phi1_0.real-V[i]*phi1_0.imag)) + mu(t)/tau_m
    

    f=l1+(Cnm(1,1,nu,d)+Cnm(1,-1,nu,d))*phih_prime(h[i],hprime)
       

    g=l1+(Cnm(1,1,nu,d)-Cnm(1,-1,nu,d))*phih_prime(h[i],hprime)
    
    c10=Cnm(1,0,nu,d)*phih_prime(h[i],hprime)
    
    K1=(f.real*U[i]-g.imag*V[i]+c10.real)
    L1=(g.real*V[i]+f.imag*U[i]+c10.imag) 
    M1=-h[i]/tau_m + J/tau_m*(A_inf +2*(U[i]*phi1_0.real-V[i]*phi1_0.imag)) +mu(t)/tau_m
    
    K2=(f.real*(U[i]+0.5*dt*K1)-g.imag*(V[i]+0.5*dt*L1)+c10.real)
    L2=(g.real*(V[i]+0.5*dt*L1)+f.imag*(U[i]+0.5*dt*K1)+c10.imag)
    M2=-(h[i]+0.5*dt*M1)/tau_m + J/tau_m*(A_inf+ 2*((U[i]+0.5*dt*K1)*phi1_0.real-(V[i]+0.5*dt*L1)*phi1_0.imag))+mu(t)/tau_m
    
    K3=(f.real*(U[i]+0.5*dt*K2)-g.imag*(V[i]+0.5*dt*L2)+c10.real)
    L3=(g.real*(V[i]+0.5*dt*L2)+f.imag*(U[i]+0.5*dt*K2)+c10.imag)
    M3=-(h[i]+0.5*dt*M2)/tau_m +J/tau_m*(A_inf +2*((U[i]+0.5*dt*K2)*phi1_0.real-(V[i]+0.5*dt*L2)*phi1_0.imag))+mu(t)/tau_m
    
    
    K4=(f.real*(U[i]+dt*K3)-g.imag*(V[i]+dt*L3)+c10.real)
    L4=(g.real*(V[i]+dt*L3)+f.imag*(U[i]+dt*K3)+c10.imag)
    M4=-(h[i]+dt*M3)/tau_m +J/tau_m*(A_inf + 2*((U[i]+dt*K3)*phi1_0.real-(V[i]+dt*L3)*phi1_0.imag))+mu(t)/tau_m
    
    
    U[i+1]=U[i]+dt/6*(K1+2*K2+2*K3+K4)
    V[i+1]=V[i]+dt/6*(L1+2*L2+2*L3+L4)
    h[i+1]=h[i]+dt/6*(M1+2*M2+2*M3+M4)
         
    
    A[i]=A_inf+2*(U[i]*phi1_0.real-V[i]*phi1_0.imag)
```

## integral formulation

```python
dt=0.001

t_max=300
N_sim=int(t_max/dt)
tsim_num=[i*dt for i in range(N_sim)]

h=np.zeros((N_sim+1,1))
h[0]=h_0

A_wc=np.zeros((N_sim,1))
A_wc[0]=1/dt


for i in range(N_sim-1):
    
    t=i*dt
    
    jstart=int(np.floor((t-d)/dt))
    
    if jstart<0:
        jstart=0
    
      
    h[i+1]=h[i]+dt/tau_m*(-h[i]+mu(i*dt)+J*A_wc[i])
    
    nu=nu_max/(1+np.exp(-beta*(h[i+1]-h_0)))
    
    A_wc[i+1]=nu*(1-np.sum(A_wc[jstart:i+1]*dt))
```

### Approximation of the eigenvalue

```python
def Cv(nu,d):
    return 1/(d*nu+1)

def R(nu,d):
    return nu/(d*nu+1)

def eigenvalue_app(r,cv2):
    return -2*np.pi**2*cv2*r + 2*np.pi*r*1j

def eigenvalue_app2(r,cv2):
    return r*(1-np.sqrt(1-4*np.pi*cv2*1j))/cv2

def eigenvalue(n,nu,d):
    return (1/d)*sc.special.lambertw(d*nu*np.exp(nu*d), k=n)-nu
```

FIX RATE R

```python
def kumulant(nu,d,n):
    m1=d+1/nu
    m2=(d*nu*(d*nu+2)+2)/nu**2
    m3=(d*nu*(d*nu*(d*nu+3)+6)+6)/nu**3
    m4=(d*nu*(d*nu*(d*nu*(d*nu+4)+12)+24)+24)/nu**4
    
    
    k1=m1
    k2=m2-m1**2
    k3=m3-3*m2*m1+2*m1**3
    k4=m4-4*m3*m1-3*m2*m2+12*m2*m1**2-6*m1**4

    
    if n==2:
        return [0.5*k2,-k1,2*np.pi*1j]
    elif n==3:
        return [-k3/6,0.5*k2,-k1,2*np.pi*1j]
    elif n==4:
        return [+1/24*k4,-k3/6,0.5*k2,-k1,2*np.pi*1j]
    else:
        print('error')
        
```
